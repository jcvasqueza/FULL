 La factorización matricial SVD se expresa como:


X=UΣV^T


Aquí está cómo podrías escribir esto en LaTeX:

\[ X = U \Sigma V^T \] codigo LaTex

Donde:

X es tu matriz original,

U es la matriz de vectores singulares izquierdos,

Σ es la matriz diagonal de valores singulares,

V^T es la matriz transpuesta de los vectores singulares derechos.


Además, en tu código anterior, mencionaste que estás utilizando el resultado de la factorización SVD para reconstruir la matriz original. Esto se hace multiplicando U,Σ, y V^T de nuevo:

Xreconstruida = UΣV^T

En LaTeX, esto se expresaría como:

\[ X_{\text{reconstruida}} = U \Sigma V^T \]



```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD

# Supongamos que 'data' es tu matriz original con 'nivel' como índice

# Paso 1: Se crea el modelo con el número correcto de componentes
modelo_svd = TruncatedSVD(n_components=data.shape[1], random_state=123)

# Pasos 2-3: Se ajusta el modelo y se pronostica
USigma = modelo_svd.fit_transform(data)

# Paso 4: Se evalúa
Sigma = np.diag(modelo_svd.singular_values_)

# Aplicar un umbral a los valores singulares
umbral = 1e-10
Sigma[Sigma < umbral] = 0

# Evitar la división por cero
U = np.divide(USigma, Sigma, out=np.zeros_like(USigma), where=(Sigma != 0))
VT = modelo_svd.components_

# Reconstruir la matriz original
data_pred = U @ Sigma @ VT

# Calcular la diferencia entre la matriz original y la matriz reconstruida
diferencia = data - data_pred

# Calcular la norma de Frobenius del error
error_frobenius = np.linalg.norm(diferencia, 'fro')

print(f"Error de Frobenius: {error_frobenius}")
```
